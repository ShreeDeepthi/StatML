```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Shruthi', 'Deepthi', 'Architha', 'Hasini', 'Sridhar', 'Sreekumar', 'Ronith', 'Abhilash', 'Sahithya', 'Akshaya', 'Preethi', 'Harshini', 'Varshitha', 'Varshith', 'Vishal', 'Rahul', 'Madhavi', 'Sreenidhi', 'Hariharan', 'BanuReddy'],
    'Percentage': [90.4, 93.7, 73.0, 92.7, 75.0, 80.0, 76.0, 95.0, 97.2, 95.5, 76.6, 91.0, 94.0, 95.0, 85.0, 80.0, 74.1, 97.4, 75.0, 95.0]
}

df = pd.DataFrame(data)

std_dev = df['Percentage'].std()

data_range = df['Percentage'].max() - df['Percentage'].min()

Q1 = df['Percentage'].quantile(0.25)
Q3 = df['Percentage'].quantile(0.75)
IQR = Q3 - Q1

print(f"Standard Deviation: {std_dev}")
print(f"Range: {data_range}")
print(f"Interquartile Range (IQR): {IQR}")
```

```python
import pandas as pd
from scipy.stats import skew, kurtosis

data = {
    'Name': ['Shruthi', 'Deepthi', 'Architha', 'Hasini', 'Sridhar', 'Sreekumar', 'Ronith', 'Abhilash', 'Sahithya', 'Akshaya', 'Preethi', 'Harshini', 'Varshitha', 'Varshith', 'Vishal', 'Rahul', 'Madhavi', 'Sreenidhi', 'Hariharan', 'BanuReddy'],
    'Percentage': [90.4, 93.7, 73.0, 92.7, 75.0, 80.0, 76.0, 95.0, 97.2, 95.5, 76.6, 91.0, 94.0, 95.0, 85.0, 80.0, 74.1, 97.4, 75.0, 95.0]
}

df = pd.DataFrame(data)

std_dev = df['Percentage'].std()

data_range = df['Percentage'].max() - df['Percentage'].min()

Q1 = df['Percentage'].quantile(0.25)
Q3 = df['Percentage'].quantile(0.75)
IQR = Q3 - Q1

skewness = skew(df['Percentage'])

kurt = kurtosis(df['Percentage'])

print(f"Standard Deviation: {std_dev}")
print(f"Range: {data_range}")
print(f"Interquartile Range (IQR): {IQR}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```
